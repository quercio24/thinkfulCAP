```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.seterr(divide='ignore', invalid='ignore')

# three different data sets, an occurrence, income, and population
data_occ = pd.read_csv('/Users/quarkoli/PycharmProjects/pythonWork/datasets/UNdata.csv')
df3 = pd.read_csv('/Users/quarkoli/PycharmProjects/pythonWork/datasets/UNincome.csv')
data_pop = pd.read_csv('/Users/quarkoli/PycharmProjects/pythonWork/datasets/UNpop.csv')
```
