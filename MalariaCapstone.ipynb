```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
from IPython.core.pylabtools import figsize

np.seterr(divide='ignore', invalid='ignore')
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['figure.dpi'] = 100
figsize(6,4)
# Plotting the mean occurrence of malaria from distribution with plus one st. deviation 
df3 = pd.read_csv('/Users/quarkoli/PycharmProjects/pythonWork/datasets/UNincome.csv')
df2 = pd.read_csv('/Users/quarkoli/PycharmProjects/pythonWork/datasets/UNpop.csv')
df = pd.read_csv('/Users/quarkoli/PycharmProjects/pythonWork/datasets/UNdata.csv')
# Start a new cell 
```

```python

mean_value = np.mean(df['Value'])
print('The mean malaria occurrence for all of the data is', round(mean_value,2))
std_value = np.std(df['Value'])
print('Standard deviation is',round(std_value,2))

# Plotting distribution of all values in a histogram 
plt.hist(df['Value'], color='blue', edgecolor='black',
         bins=int(180/9.25))
plt.ylim(0, 200)
plt.axvline(mean_value, color='r', linestyle='solid', linewidth=2)
plt.axvline(mean_value + std_value, color='black', linestyle='dashed', linewidth=2)

# Adding labels
plt.title("Histogram of documented malaria cases from 2002-2012")
plt.xlabel('Cases of malaria annually (thousands)')
plt.ylabel('Frequency of cases')
plt.xlim(0, 3.0e+06)
plt.ylim(0,1000)
plt.grid(axis='y', alpha=0.75)
plt.text(1.0e+06,420, r'mean= 1.86e+05, '
                      r'std= +4.69e+05 ')
plt.show();
```

```python
# # CREATING PLOTS FOR SIX COUNTRIES THAT I CHOSE

# # Making a separate array for each country, then using this as plotting data
figsize(8,5)
x1 = df[df['Country or Area'] == 'Angola']
x2 = df[df['Country or Area'] == 'Burundi']
x3 = df[df['Country or Area'] == 'Democratic Republic of the Congo']
x4 = df[df['Country or Area'] == 'India']
x5 = df[df['Country or Area'] == 'United Republic of Tanzania']
x6 = df[df['Country or Area'] == 'Yemen']

#### Overlaying line plots, showing cases of malaria for all 5 countries over time  ####

plt.title('Malaria cases among random countries')
plt.ylabel('Annual occurrences')
plt.xlabel('Years')
plt.plot(x1['Year(s)'],x1['Value'],label='Angola', color='violet')
plt.plot(x2['Year(s)'],x2['Value'],label='Burundi', color='green')
plt.plot(x3['Year(s)'],x3['Value'],label='Democratic Republic of the Congo', color='blue')
plt.plot(x4['Year(s)'],x4['Value'],label='India', color='orange')
plt.plot(x5['Year(s)'],x5['Value'],label='United Republic of Tanzania', color='red')
plt.plot(x6['Year(s)'],x6['Value'],label= 'Yemen' , color='indigo')
plt.legend()
plt.show()

# maximum occurrence of malaria was Congo, 2012
max_malaria = df['Value'].max()
print('The max malaria occurrence was', max_malaria) # around 186k cases annually 
```

```python
figsize(8,7)
mean_income = np.mean(df3['Value'])
std_income = np.std(df3['Value'])
print('The mean income per capita is', round(mean_income,2))
print('The standard deviation per capita is', round(std_income,2))

plt.hist(df3['Value'], color='blue', edgecolor='black',
         bins='auto')
plt.title("Histogram of income per capita from 2002-2012")
plt.axvline(mean_income, color='r', linestyle='solid', linewidth=2)
plt.axvline(mean_income + std_income, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('Annual Income per capita (thousands)')
plt.ylabel('Frequency of income')
plt.xlim(0, 80000)
plt.ylim(0,1500)
plt.grid(axis='y', alpha=0.75)

plt.show();
```
